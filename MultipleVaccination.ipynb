```python
import numpy as np


#l_0 = 0.01
#l_1 = 0.001
#theta_l = 10

def rr(rr_0,theta1,rr_1,t):
  if t < theta1:
    return rr_0
  else:
    return rr_1
  #elif t < theta3:
  #  return rr_2
  #else:
  #  return rr_3

#def l(l_0,l_1,theta,t):
#  if t < theta:
#    return l_0
#  else:
#    return l_1

l = 1/21

def fV(V,F,S,E,I,A,R,beta1,beta2,gamma,rr_0,theta1,rr_1,l,t):
  return (1-eA)*rr(rr_0,theta1,rr_1,t)*S - eL1*(beta1*I + beta2*A)*V - l*V

def fF(V,F,S,E,I,A,R,beta1,beta2,gamma,rr_0,theta1,rr_1,l,t):
  return l*V - alpha*F - eL2*(beta1*I + beta2*A)*F

def fS(V,F,S,E,I,A,R,beta1,beta2,gamma,rr_0,theta1,rr_1,l,t):
  return -(beta1*I + beta2*A)*S + alpha*F - (1-eA)*rr(rr_0,theta1,rr_1,t)*S

def fE(V,F,S,E,I,A,R,beta1,beta2,gamma,rr_0,theta1,rr_1,l,t):
  return (beta1*I + beta2*A)*(eL1*V+eL2*F) + (beta1*I + beta2*A)*S - sigma*E

def fI(V,F,S,E,I,A,R,beta1,beta2,gamma,rr_0,theta1,rr_1,l,t):
  return w*sigma*E - (gamma + delta)*I

def fA(V,F,S,E,I,A,R,beta1,beta2,gamma,rr_0,theta1,rr_1,l,t):
  return (1-w)*sigma*E - gamma*A

def fR(V,F,S,E,I,A,R,beta1,beta2,gamma,rr_0,theta1,rr_1,l,t):
  return gamma*(I + A) 
```

```python
def rr(rr_0,rr_1,theta,t):
  if t < theta:
    return rr_0
  else:
    return rr_1

def l(l_0,l_1,theta,t):
  if t < theta:
    return l_0
  else:
    return l_1

# PR = [rr_10,rr_11,theta_rr1,rr_20,rr_21,theta_rr2,rr_30,rr_31,theta_rr3]
# PL = [l_10,l_11,theta_l1,l_30,l_31,theta_l3]

def fV1(V1,F1,V2,V3,F3,S,E,I,A,R,beta1,beta2,gamma,PR,PL,t):
  rr_0 = PR[0]
  rr_1 = PR[1]
  theta_rr = PR[2]
  l_0 = PL[0]
  l_1 = PL[1]
  theta_l = PL[2]

  return (1-e1A)*rr(rr_0,rr_1,theta_rr,t)*S - e1L1*(beta1*I + beta2*A)*V1 - l(l_0,l_1,theta_l,t)*V1

def fV2(V1,F1,V2,V3,F3,S,E,I,A,R,beta1,beta2,gamma,PR,PL,t): # Espacial
  rr_0 = PR[3]
  rr_1 = PR[4]
  theta_rr = PR[5]
  return (1-e2A)*rr(rr_0,rr_1,theta_rr,t)*S - e2L1*(beta1*I + beta2*A)*V2 - alpha2*V2

def fV3(V1,F1,V2,V3,F3,S,E,I,A,R,beta1,beta2,gamma,PR,PL,t):
  rr_0 = PR[6]
  rr_1 = PR[7]
  theta_rr = PR[8]
  l_0 = PL[3]
  l_1 = PL[4]
  theta_l = PL[5]
  return (1-e3A)*rr(rr_0,rr_1,theta_rr,t)*S - e3L1*(beta1*I + beta2*A)*V3 - l(l_0,l_1,theta_l,t)*V3


def fF1(V1,F1,V2,V3,F3,S,E,I,A,R,beta1,beta2,gamma,PR,PL,t):
  rr_0 = PR[0]
  rr_1 = PR[1]
  theta_rr = PR[2]
  l_0 = PL[0]
  l_1 = PL[1]
  theta_l = PL[2]  
  return l(l_0,l_1,theta_l,t)*V1 - alpha1*F1 - e1L2*(beta1*I + beta2*A)*F1

def fF3(V1,F1,V2,V3,F3,S,E,I,A,R,beta1,beta2,gamma,PR,PL,t):
  rr_0 = PR[6]
  rr_1 = PR[7]
  theta_rr = PR[8]
  l_0 = PL[3]
  l_1 = PL[4]
  theta_l = PL[5] 
  return l(l_0,l_1,theta_l,t)*V3 - alpha3*F3 - e3L2*(beta1*I + beta2*A)*F3

def fS(V1,F1,V2,V3,F3,S,E,I,A,R,beta1,beta2,gamma,PR,PL,t):
  rr_10 = PR[0]
  rr_11 = PR[1]
  theta_rr1 = PR[2]
  rr_20 = PR[3]
  rr_21 = PR[4]
  theta_rr2 = PR[5]
  rr_30 = PR[6]
  rr_31 = PR[7]
  theta_rr3 = PR[8]    
  ret = -(beta1*I + beta2*A)*S 
  ret += alpha1*F1 - (1-e1A)*rr(rr_10,rr_11,theta_rr1,t)*S
  ret += alpha2*V2 - (1-e2A)*rr(rr_20,rr_21,theta_rr2,t)*S
  ret += alpha3*F3 - (1-e3A)*rr(rr_30,rr_31,theta_rr3,t)*S
  return ret

def fE(V1,F1,V2,V3,F3,S,E,I,A,R,beta1,beta2,gamma,PR,PL,t):
  return (beta1*I + beta2*A)*(e1L1*V1+e1L2*F1 + e2L1*V2 + e3L1*V3+e3L2*F3) + (beta1*I + beta2*A)*S - sigma*E

def fI(V1,F1,V2,V3,F3,S,E,I,A,R,beta1,beta2,gamma,PR,PL,t):
  return w*sigma*E - (gamma + delta)*I

def fA(V1,F1,V2,V3,F3,S,E,I,A,R,beta1,beta2,gamma,PR,PL,t):
  return (1-w)*sigma*E - gamma*A

def fR(V1,F1,V2,V3,F3,S,E,I,A,R,beta1,beta2,gamma,PR,PL,t):
  return gamma*(I + A) 
```

```python
def Loss(beta1,beta2,gamma,rr_10,rr_11,theta_rr1,rr_20,rr_21,theta_rr2,rr_30,rr_31,theta_rr3,l_10,l_11,theta_l1,l_30,l_31,theta_l3):
  m = 1
  h1 = 1 / m

  v1 = V1D[0] 
  v1c = V1D[0]
  f1 = F1D[0]
  f1c = F1D[0]
  v2 = V2D[0] 
  v2c = V2D[0]
  v3 = V3D[0] 
  v3c = V3D[0]
  f3 = F3D[0]
  f3c = F3D[0]
  #s = SD[0]
  e = E0
  i = ID[0]
  a = (1-w)*ID[0]
  r = RD[0]
  d = DD[0]
  s = 1 - v1 - f1 - v2 - v3 - f3 - e - i - a - r - d 

  #print(s)

  V1 = [v1]
  V1C = [v1c]
  F1 = [f1]
  F1C = [f1c]
  V2 = [v2]
  V2C = [v2c]
  V3 = [v3]
  V3C = [v3c]
  F3 = [f3]
  F3C = [f3c]


  S = [s]
  E = [e]
  I = [i]
  A = [a]
  R = [r]
  D = [d]

  PR = [rr_10,rr_11,theta_rr1,rr_20,rr_21,theta_rr2,rr_30,rr_31,theta_rr3]
  PL = [l_10,l_11,theta_l1,l_30,l_31,theta_l3]


  for tempo in range(1,len(TD)*m): #10000

    t = tempo / m

    #print(v1+v2+s+e+i+a+h+u+r+d)
    v11 = v1 + h1*fV1(v1,f1,v2,v3,f3,s,e,i,a,r,beta1,beta2,gamma,PR,PL,t)
    v1c1= v1c + h1*rr(rr_10,rr_11,theta_rr1,t)*s
    f11 = f1 + h1*fF1(v1,f1,v2,v3,f3,s,e,i,a,r,beta1,beta2,gamma,PR,PL,t)
    f1c1= f1c + l(l_10,l_11,theta_l1,t)*(v1c1-f1c)#v
    v21 = v2 + h1*fV2(v1,f1,v2,v3,f3,s,e,i,a,r,beta1,beta2,gamma,PR,PL,t)
    v2c1= v2c + h1*rr(rr_20,rr_21,theta_rr2,t)*s
    v31 = v3 + h1*fV3(v1,f1,v2,v3,f3,s,e,i,a,r,beta1,beta2,gamma,PR,PL,t)
    v3c1= v3c + h1*rr(rr_30,rr_31,theta_rr3,t)*s
    f31 = f3 + h1*fF3(v1,f1,v2,v3,f3,s,e,i,a,r,beta1,beta2,gamma,PR,PL,t)
    f3c1= f3c + l(l_30,l_31,theta_l3,t)*(v3c1-f3c)#v
    s1 = s + h1*fS(v1,f1,v2,v3,f3,s,e,i,a,r,beta1,beta2,gamma,PR,PL,t)
    e1 = e + h1*fE(v1,f1,v2,v3,f3,s,e,i,a,r,beta1,beta2,gamma,PR,PL,t)
    i1 = i + h1*fI(v1,f1,v2,v3,f3,s,e,i,a,r,beta1,beta2,gamma,PR,PL,t)
    a1 = a + h1*fA(v1,f1,v2,v3,f3,s,e,i,a,r,beta1,beta2,gamma,PR,PL,t)
    r11 = r + h1*fR(v1,f1,v2,v3,f3,s,e,i,a,r,beta1,beta2,gamma,PR,PL,t)
    d1 = d + h1*delta*i


    v1 = v11 + 0
    v1c = v1c1 + 0
    f1c = f1c1 + 0
    f1 = f11 + 0
    v2 = v21 + 0
    v2c = v2c1 + 0  
    v3 = v31 + 0
    v1c = v3c1 + 0
    f3c = f3c1 + 0
    f3 = f31 + 0  
    s = s1 + 0
    e = e1 + 0
    i = i1 + 0
    a = a1 + 0
    r = r11 + 0
    d = d1 + 0

    V1.append(v1)
    V1C.append(v1c)
    F1.append(f1)
    F1C.append(f1c)
    V2.append(v2)
    V2C.append(v2c)
    V3.append(v3)
    V3C.append(v3c)
    F3.append(f3)
    F3C.append(f3c)
    S.append(s)
    E.append(e)
    I.append(i)
    A.append(a)
    R.append(r)
    D.append(d)


  I = np.array(I)
  R = np.array(R)
  V1C = np.array(V1C)
  F1C = np.array(F1C)
  V2C = np.array(V2C)
  V3C = np.array(V3C)
  F3C = np.array(F3C)

  IDnp = np.array(ID)
  RDnp = np.array(RD)
  V1Dnp = np.array(V1D)
  F1Dnp = np.array(F1D)
  V2Dnp = np.array(V2D)
  V3Dnp = np.array(V3D)
  F3Dnp = np.array(F3D)

  suma = (I-IDnp)**2 + (R - RDnp)**2 + 10*(V1C - V1Dnp)**2 + 10*(F1C - F1Dnp)**2 + (V2C - V2Dnp)**2 + (V3C - V3Dnp)**2 + (F3C - F3Dnp)**2  
  suma = np.sum(suma) 
  suma += pos(V1) + pos(V2) + pos(V3) + pos(F1) + pos(F3) + pos(S) + pos(E) + pos(I) + pos(A) + pos(R) + pos(D)
  return suma
```

```python
from math import exp

#def rr(rr_0,rr_1,theta_rr,t):
#  if t < theta_rr:
#    return rr_0
#  else:
#    return rr_1

def rr(rr_0,rr_1,t):
  return rr_0*exp(rr_1*t)

def rr2(rr_0,rr_1,theta_rr,t):
  if t < theta_rr:
    return 0
  else:
    return rr_0*exp(rr_1*t)


def l(l_0,l_1,t):
  return l_0*exp(l_1*t)
  #if t < theta:
  #  return l_0
  #else:
  #  return l_1

# PR = [rr_10,rr_11,theta_rr1,rr_20,rr_21,theta_rr2,rr_30,rr_31,theta_rr3,rr_40,rr_41,theta_rr4]
# PL = [l_10,l_11,theta_l1,l_20,l_21,theta_l2,l_30,l_31,theta_l3]

def fV1(V1,F1,V2,F2,V3,F3,V4,S,E,I,A,R,beta1,beta2,gamma,PR,PL,t):
  rr_0 = PR[0]
  rr_1 = PR[1]
  l_0 = PL[0]
  l_1 = PL[1]

  return (1-e1A)*rr(rr_0,rr_1,t)*S - e1L1*(beta1*I + beta2*A)*V1 - l(l_0,l_1,t)*V1

def fV2(V1,F1,V2,F2,V3,F3,V4,S,E,I,A,R,beta1,beta2,gamma,PR,PL,t): # Espacial
  rr_0 = PR[2]
  rr_1 = PR[3]
  l_0 = PL[2]
  l_1 = PL[3]
  return (1-e2A)*rr(rr_0,rr_1,t)*S - e2L1*(beta1*I + beta2*A)*V2 - l(l_0,l_1,t)*V2

def fV3(V1,F1,V2,F2,V3,F3,V4,S,E,I,A,R,beta1,beta2,gamma,PR,PL,t):
  rr_0 = PR[4]
  rr_1 = PR[5]
  l_0 = PL[4]
  l_1 = PL[5]
  return (1-e3A)*rr(rr_0,rr_1,t)*S - e3L1*(beta1*I + beta2*A)*V3 - l(l_0,l_1,t)*V3

def fV4(V1,F1,V2,F2,V3,F3,V4,S,E,I,A,R,beta1,beta2,gamma,PR,PL,t): # Especial
  rr_0 = PR[6]
  rr_1 = PR[7]
  theta_rr = PR[8]
  return (1-e4A)*rr2(rr_0,rr_1,theta_rr,t)*S - e4L1*(beta1*I + beta2*A)*V4 - alpha4*V4

def fF1(V1,F1,V2,F2,V3,F3,V4,S,E,I,A,R,beta1,beta2,gamma,PR,PL,t):
  rr_0 = PR[0]
  rr_1 = PR[1]
  l_0 = PL[0]
  l_1 = PL[1]
  return l(l_0,l_1,t)*V1 - alpha1*F1 - e1L2*(beta1*I + beta2*A)*F1

def fF2(V1,F1,V2,F2,V3,F3,V4,S,E,I,A,R,beta1,beta2,gamma,PR,PL,t):
  rr_0 = PR[2]
  rr_1 = PR[3]
  l_0 = PL[2]
  l_1 = PL[3]  
  return l(l_0,l_1,t)*V2 - alpha2*F2 - e2L2*(beta1*I + beta2*A)*F2

def fF3(V1,F1,V2,F2,V3,F3,V4,S,E,I,A,R,beta1,beta2,gamma,PR,PL,t):
  rr_0 = PR[4]
  rr_1 = PR[5]
  l_0 = PL[4]
  l_1 = PL[5]
  return l(l_0,l_1,t)*V3 - alpha3*F3 - e3L2*(beta1*I + beta2*A)*F3

def fS(V1,F1,V2,F2,V3,F3,V4,S,E,I,A,R,beta1,beta2,gamma,PR,PL,t):
  rr_10 = PR[0]
  rr_11 = PR[1]
  rr_20 = PR[2]
  rr_21 = PR[3]
  rr_30 = PR[4]
  rr_31 = PR[5]
  rr_40 = PR[6]
  rr_41 = PR[7]
  theta_rr = PR[8]
  ret = -(beta1*I + beta2*A)*S 
  ret += alpha1*F1 - (1-e1A)*rr(rr_10,rr_11,t)*S
  ret += alpha2*F2 - (1-e2A)*rr(rr_20,rr_21,t)*S
  ret += alpha3*F3 - (1-e3A)*rr(rr_30,rr_31,t)*S
  ret += alpha4*V4 - (1-e4A)*rr2(rr_40,rr_41,theta_rr,t)*S
  return ret

def fE(V1,F1,V2,F2,V3,F3,V4,S,E,I,A,R,beta1,beta2,gamma,PR,PL,t):
  return (beta1*I + beta2*A)*(e1L1*V1+ e1L2*F1 + e2L1*V2 + e2L2*F2 + e3L1*V3+e3L2*F3 + e4L1*V4) + (beta1*I + beta2*A)*S - sigma*E

def fI(V1,F1,V2,F2,V3,F3,V4,S,E,I,A,R,beta1,beta2,gamma,PR,PL,t):
  w = PL[6]
  return w*sigma*E - (gamma + delta)*I

def fA(V1,F1,V2,F2,V3,F3,V4,S,E,I,A,R,beta1,beta2,gamma,PR,PL,t):
  w = PL[6]
  return (1-w)*sigma*E - gamma*A

def fR(V1,F1,V2,F2,V3,F3,V4,S,E,I,A,R,beta1,beta2,gamma,PR,PL,t):
  return gamma*(I + A) 
```
